```
import tensorflow as tf
import pickle
import numpy as np
```

```
model = tf.keras.models.load_model("model_songsWithTags_20000.h5")
dataset = pickle.load(open('input_spectrograms_2000.pickle', 'rb'))

dataset_X_init = []
dataset_y_init = []
for sample in dataset:
    dataset_X_init.append(sample[0])
    dataset_y_init.append(sample[1])
  
dataset_X = np.array([x[:,:431].reshape((128, 431, 1)) for x in dataset_X_init])
dataset_y = np.array(tf.keras.utils.to_categorical(dataset_y_init, 2))

```

```
dataset_X.shape
```

```
dataset_y.shape
```

```
score = model.evaluate(x=dataset_X, y=dataset_y)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```
predicted_y = np.argmax(model.predict(dataset_X), axis=-1)
```

```
predicted_y
```

```
missclassified_idx = np.where(predicted_y != np.array(dataset_y_init))[0]
```

```
misclassified_genre = np.array(dataset_y_init)[missclassified_idx]
```
